#### Define

Create a new column that selects the most trustworthy predicted breed; If all results are false, return "Cannot Predict"

#### Code

```python
#Create a new column that selects the most trustworthy predicted breed, otherwise returen "Cannot Predict"
def predict_breed(row):
    if row.p1_dog:
        return row.p1
    elif row.p2_dog:
        return row.p2
    elif row.p3_dog:
        return row.p3
    else:
        return "Cannot Predict"
image_clean['dog_breed'] = image_clean.apply(lambda row: predict_breed(row),axis=1)
```
